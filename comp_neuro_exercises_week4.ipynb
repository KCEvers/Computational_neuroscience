## exercises week 4: Kuramoto

The Kuramoto model is a mathematical model used to describe the dynamics of sets of coupled oscillators. As we have seen in the previous lecture and assignment, if a neuron bifurcates towards a stable limit cycle (it spikes regularly) it's dynamics can be described by the phase alone. Under this assumption, a neuron is essentially a phase model or, in other words, an oscillator. Making a neural network a population of interacting limit-cycle oscillators. As such, Kuramoto's simplification comes in handy when we want to describe the dynamics of large groups of neurons, keeping the analysis tractable.

We will see that large groups of coupled oscillators tend to synchronise their phases, even though their individual frequencies may be wide-ranging. Moreover, we find that these populations can undergo the temporal analog of a phase transition.
